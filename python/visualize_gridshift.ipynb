```python
import numpy as np
import matplotlib.pylab as plt
```

```python
data = []

for grid_size in [0.01, 0.04, 0.07, 0.1, 0.13, 0.16]:
    data.append([])
    for n in [10000, 50000, 100000, 500000, 1000000]:
        with open(f"../logs/gridShift/N{n}_GS{grid_size:.06f}.csv", "r") as f:
            data[-1].append(np.mean(list(map(lambda line: float(line.strip()), f.readlines()))))

data = np.array(data)
data.shape
```

```python
plt.plot(data[0])
plt.show()
```

```python
points = []
centroids1 = []
centroids2 = []

with open("../logs/gridShift/test.log", "r") as f:
    for line in f:
        split = line.split(",")
        if split[0] == "point":
            point = float(split[1]), float(split[2])
            points.append(point)
        elif split[0] == "cluster1":
            point = float(split[1]), float(split[2])
            centroids1.append(point)
        elif split[0] == "cluster2":
            point = float(split[1]), float(split[2])
            centroids2.append(point)

points = np.array(points)
centroids1 = np.array(centroids1)
centroids2 = np.array(centroids2)

points.shape, centroids1.shape, centroids2.shape
```

```python
plt.figure(figsize=(8, 8), dpi=120)
plt.scatter(points[:, 0], points[:, 1], s=0.5)
# plt.scatter(centroids1[:, 0], centroids1[:, 1])
plt.scatter(centroids2[:, 0], centroids2[:, 1], c="red")
plt.show()
```
